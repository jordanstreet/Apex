```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
```

```python
# Configure plotting style
plt.rcParams['figure.figsize'] = [16, 10]
```

```python
# Load training data
data   = pickle.load(open('trainingData.db', 'rb'))
images = np.array(data['images'])
labels = np.array(data['labels'])

# Flatten images
N      = len(images)
images = images.reshape(N, -1)

# Create training and testing data
imagesTrain, imagesTest, labelsTrain, labelsTest = train_test_split(images, labels)
```

```python
# Create and train SVM classifier
clf = SVC()
clf.fit(imagesTrain, labelsTrain)

# Evaluate performance
labelsPredicted = clf.predict(imagesTest)

# Print performance
print(f"Classification report for classifier {clf}:\n"
      f"{classification_report(labelsTest, labelsPredicted)}\n")

# Display confusion matrix
plot_confusion_matrix(clf, imagesTest, labelsTest)

# Save classifier
pickle.dump(clf, open('apexOCR.db', 'wb'))
```
